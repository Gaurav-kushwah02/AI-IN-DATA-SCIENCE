```python
#important libraries
import pandas as pd
```

```python
#Upload DataSet for cleaning
data=pd.read_csv('/content/Messy_Employment_India_Dataset.csv')
```

# MISSING VALUE

```python
data.isnull()    #to check missing value, true is missing false if data is filled.
```

```python
data.isnull().sum()         #sum up all the null space
```

# HANDLING MISSING VALUES
### fillna()..this func for filling value inside it at empty space

```python
#replacing all null space by finding mean of it.
data['Years of Experience'] = data['Years of Experience'].fillna(data['Years of Experience'].mean())
data
```

```python
#replacing null value by word
data['Status']=data['Status'].fillna('UNKNOWN')
data
```

# REMOVE DUPLICATES IN DATASET

```python
#it removes the duplicate rows
data.drop_duplicates()
data
```

# Check And Fix Data type

```python
#o check data type of every group
data.dtypes
```

```python
#To change the data type
data['Age Group'] = data['Age Group'].astype(int)

```

# Rename The Columns

```python
#it will change the column heading, We can change multiple name at a time just by using "," in curly braces
data=data.rename(columns={'Age Group':'Age'})
data
```

# Handle Outliers (Simple Filtering by using conditions)

```python
#it will sort data based on condition
data_sort=data[data['Monthly Salary (INR)']<10000]
data_sort
```

# Clean Text Data


```python
#Removes extra spaces from the beginning and end
data['Education'] = data['Education'].str.strip()
data
```

```python
#Converts all city names to lowercase letters.
data['Location'] = data['Location'].str.lower()
data
```

# DATA  VALIDATION

```python
#it will show data of those have YOE more than 10 year
data=data[data['Years of Experience']>10]
data
```

```python
#it will show data acc to condition
data = data[(data['Age'] > 0) & (data['Age'] <= 120)]
```

# To save the clean or modified file

```python
data.to_csv("cleared data.csv",index=False)
```
