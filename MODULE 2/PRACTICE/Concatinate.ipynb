# CONCATINATE

```python
import pandas as pd
```

### Concatinating in Series

```python
#Creating series
sr1=pd.Series([1,2,3])
sr2=pd.Series([4,5,6])

#concatinating the data
data=pd.concat([sr1,sr2])
data
```

### Concatinating in DataFrame

```python
#create DataFrames
dataf1=pd.DataFrame({'A':[1,2,3,4],'B':[11,21,31,41]})
dataf2=pd.DataFrame({'A':[1,2,3,5],'C':[10,20,30,40]})
#concatinating the data
data=pd.concat([dataf1,dataf2])
data
```

```python
#concatinating the data
sr_data=pd.concat([sr1,sr2])
sr_data
```

# JOIN On Concat

```python
import pandas as pd

# Create first DataFrame

df1 = pd.DataFrame({
    "id": [1, 2, 3],
    "name": ["A", "B", "C"]
})

df2 = pd.DataFrame({
    "id": [2, 3, 4],
    "score": [80, 90, 100]
})

# Join using concat (set key as index)
result = pd.concat([df1.set_index("id"), df2.set_index("id")],axis=1)
result                                                       # Display result

#Ye sirf index ke basis par columns jodta hai
#Default behavior = outer join
```

```python
#Outer join (default)
result_outer = pd.concat([df1.set_index("id"), df2.set_index("id")],axis=1,join="outer")
result_outer

#Union of keys
#Agar kisi df me key missing ho  NaN
```

```python
#Inner join (only matching ids)
result_inner = pd.concat([df1.set_index("id"), df2.set_index("id")],axis=1,join="inner")
result_inner
#Common keys only
#Jo dono me present ho wahi aayega
```
