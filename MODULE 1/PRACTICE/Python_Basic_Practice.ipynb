```

def calculation():
 num1=float(input("Enter number: "))     #taking 1st number
 op=input("Enter operation: + - * / ")   #select operation
 num2=float(input("Enter number: "))     #taking 2nd number

 if op == "+":                          #checking condition for operation
   print("Sum of numbers: ",num1+num2)
 elif op == "-":
   print("Subtraction of numbers: ",num1-num2)
 elif op == "*":
   print("Multiplication of numbers: ",num1*num2)
 elif op == "/":
   if op == 0:
     print("Error: Division by zero")              #error checking if diving by 0
 else:
    print("Division of numbers: ",num1/num2)

calculation()
```
