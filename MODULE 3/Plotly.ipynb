# Data Visualization with Plotly

```python
#important lib
import matplotlib.pyplot as plt
import pandas as pd
import plotly.express as px
```

```python
Ind_data=px.data.gapminder().query("country=='India'")   #using sql query to short data
Ind_data
```

# Bar Chart

```python
#Bar chart
bar=px.bar(Ind_data,x="year",y="pop",title="Population of India",labels={"pop":"Population","year":"Year"},color="pop",
           hover_data=["lifeExp","gdpPercap"],text="pop")
bar
```

```python
bar.update_traces(texttemplate='%{text:.2s}', textposition='inside')
```

```python
bar.update_layout(uniformtext_minsize=10, xaxis_tickangle=-45,yaxis_tickangle=-45,template="plotly_dark")
```

# Line Plot

```python
line=px.line(Ind_data,x="year",y="pop",title="Population of India Over The Year",hover_data=["lifeExp","gdpPercap"],text="pop")
line
```

# Scatter Plot

```python
scatter_plot=px.scatter(Ind_data,x="gdpPercap",y="lifeExp",size="pop",title="GDP And Life",hover_data=["lifeExp","gdpPercap"],text="pop")
scatter_plot
```

```python
scatter_plot.update_traces(texttemplate='%{text:.2s}')
```

# Pie Chart

```python
#Pie Chart
pie_chart=px.pie(Ind_data,values="pop",names="year",hole=0.1)
pie_chart
```

# Histogram

```python
#Histogram
hist=px.histogram(Ind_data,x="year",y="pop")
hist
```

```python
#heat map
heatmap=px.density_heatmap(Ind_data,x="year",y="pop")
heatmap
```

# Box Plot

```python
#Boxplot
boxplot=px.box(Ind_data,y="pop",points="all",color_discrete_sequence=["green"])
boxplot
```

```python
boxplot.update_traces(boxmean=True,jitter=0.5,marker=dict(size=15,opacity=0.5))
```
