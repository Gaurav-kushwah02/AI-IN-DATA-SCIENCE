## **Ques.** **Mention each data point names with the help of loops.**


```python
import matplotlib.pyplot as plt

# Data
names = ['Rohan', 'Amar', 'Gaurav', 'Ankit']
age = [12, 17, 19, 28]
plt.plot(names,age,color="r", marker="o")
# Create scatter plot
plt.scatter(names, age, color='blue')

# Loop to label each data point
for i in range(len(names)):
    plt.text(names[i], age[i], names[i], fontsize=10, ha='left', va='bottom')

# Add labels and title
plt.xlabel('Names')
plt.ylabel('Age')
plt.title("Details")

# Show plot
plt.show()
```

## **Ques. Can we draw 2 lines in a single graph**

```python
# Data for lines
x = ['Ram','Rohan','Karan','Sumit']
y = [10, 20, 15, 25]
z= [12, 17, 19, 28]

# Ploting both lines
plt.plot(x, y,color='y', marker='o')
plt.plot(x, z,color='g', marker='s')

# Add labels, title
plt.xlabel('Age')
plt.ylabel('Name')
plt.title('Data')

# Show plot
plt.show()
```

# **Add legend**

```python
# Data for lines
x = ['Ram','Rohan','Karan','Sumit']
y = [10, 20, 15, 25]
z= [12, 17, 19, 28]

# Ploting both lines
plt.plot(x, y,color='y', marker='o',label="age")
plt.plot(x, z,color='g', marker='s',label="age")

# Add labels, title
plt.xlabel('Name')
plt.ylabel('Age')
plt.title('Details')
#for legend we need to add label while ploting lines
plt.legend()

# Show plot
plt.show()
```

# **Draw subplot for line chart**

```python
# Data
x = ['Ram','Rohan','Karan','Sumit']
y = [10, 20, 15, 25]
z = [12, 17, 19, 28]

# Create 2 subplots (1 row, 2 columns)
plt.subplot(1, 2, 1)
plt.plot(x, y, color='b', marker='o')
plt.title('Line Chart Y')
plt.xlabel('Name')
plt.ylabel('Age')

plt.subplot(1, 2, 2)
plt.plot(x, z, color='g', marker='s')
plt.title('Line Chart Z')
plt.xlabel('Name')
plt.ylabel('Age')

# Adjust layout
plt.tight_layout()

# Show plot
plt.show()
```
